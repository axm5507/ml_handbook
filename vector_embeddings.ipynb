```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
sentences = ['Messi is the GOAT! Soccer 4 Lyfe!', 'Ronaldo is the best football player of all time.', 'Julius Caesar was a great emperor.', 'History is fascinating!']
embeddings = model.encode(sentences)
print(embeddings)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(embeddings)

print("Cosine Similarity Matrix:")
print(similarity_matrix)
```

```python

similarity_threshold = 0.25 # e.g., 50% similarity

print(f"\nSentence groups with cosine similarity >= {similarity_threshold*100:.0f}%:\n")

already_grouped_indices = set()

for i in range(len(sentences)):
    if i not in already_grouped_indices:
        current_group = [i]
        for j in range(i + 1, len(sentences)):
            if similarity_matrix[i][j] >= similarity_threshold:
                current_group.append(j)

        if len(current_group) > 1:
            print(f"Group:")
            for idx in current_group:
                print(f"  - Sentence '{sentences[idx]}' (Index: {idx})")

            # Print similarity between members of the group
            for k in range(len(current_group)):
                for l in range(k + 1, len(current_group)):
                    idx1 = current_group[k]
                    idx2 = current_group[l]
                    similarity = similarity_matrix[idx1][idx2]
                    print(f"    Similarity between sentence {idx1} and {idx2}: {similarity*100:.2f}%")

            for idx in current_group:
                already_grouped_indices.add(idx)
            print("\n" + "-"*50 + "\n")

if not already_grouped_indices:
    print("No groups found above the specified similarity threshold.")
```
